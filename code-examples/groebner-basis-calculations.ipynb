```python
import sympy as sp 
```

```python
[x,y,z,w]=sp.symbols('x y z w')
```

```python
p1 = x**2 + 2*x*y*z - 2 * x - 3 * y - 5
p2 = y**2 - 2 * x* y * z - x - 2*y - 7
p3 = x**2 - w**2
p4 = x + 2 * w - 3 * y + z
```

```python
G = sp.groebner([p1, p2, p3, p4], [x, y, z, w], domain='QQ', order='grlex')
```

```python
print(G)
```

```python
G = sp.groebner([p1, p2, p3, p4], [x, y, z, w], domain='QQ', order='lex')
```

```python
print(G)
```
